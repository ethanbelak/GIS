  - Data important and management
    - This part was easier for me because I am usually very organized with my files, but if it wasn't organized it would be extremely difficult to accomplish. Keeping things in separate and distinctly labelled files helped a lot.
  - Map creation and annotation
    - This part was a little harder for me as it took me some time to learn and adjust to the new software.
